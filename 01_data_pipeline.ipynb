```python
import pandas as pd 
import numpy as np 

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
raw_path = "../data/raw/superstore.csv"

df_raw= pd.read_csv(raw_path)

print('Shape: ', df_raw.shape)
df_raw.head()
```

```python
df_raw.info()
```

```python
df_raw.columns.tolist()
```

```python
for col in df_raw.columns:
    print(col)
```

```python
df_raw= df_raw.drop(columns=['记录数'],errors='ignore')
df_raw.columns.to_list()
```

```python
df_raw.columns = (
    df_raw.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('-', '_')
        .str.replace('/', '_')
        .str.replace('.', '_')   # <-- important fix
)

df_raw.columns.tolist()
```

```python
df_raw.duplicated().sum()
```

```python
df_raw.duplicated(subset=['order_id','product_id']).sum()
```

```python
df_clean= (
    df_raw
    .groupby(['order_id', 'product_id'],as_index=False)
    .agg({
        'sales': 'sum',
        'profit': 'sum',
        'quantity': 'sum',
        'discount': 'mean',
        'order_date': 'first',
        'ship_date': 'first',
        'customer_id': 'first',
        'customer_name': 'first',
        'segment': 'first',
        'category': 'first',
        'sub_category': 'first',
        'region': 'first',
        'market': 'first'
    })
)
```

```python
df_clean.shape
df_clean.duplicated(subset=['order_id','product_id']).sum()
```

```python
date_cols= ['order_date', 'ship_date']

for cols in date_cols:
    df_clean[cols]= pd.to_datetime(df_clean[cols], errors='coerce')

df_clean[date_cols].dtypes
```

```python
#checking if another columns need hcanged data types 
df_clean.dtypes
```

```python
cat_cols=[
    'segment',
    'category', 
    'sub_category', 
    'region', 
    'market'
]

for col in cat_cols: 
    df_clean[col]= df_clean[col].astype('category')
```

```python
df_clean[cat_cols].dtypes
```

```python
print(df_clean.shape)
print(df_clean.isnull().sum().sort_values(ascending=False).head(10))

print("Duplicates (order_id, product_id):", 
      df_clean.duplicated(subset=['order_id','product_id']).sum())
```

```python
df_clean.to_csv("../data/processed/superstore_clean.csv",index=False)

print("Exported")
```
